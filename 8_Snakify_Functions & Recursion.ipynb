```python
from math import sqrt
def distance(x1, y1, x2, y2):
  return sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
x1 = float(input())
x2 = float(input())
y1 = float(input())
y2 = float(input())
print(distance(x1, x2, y1, y2))
```

```python
def power(a, n):
  res = 1
  for i in range(abs(n)):
    res *= a
  if n >= 0:
    return res
  else:
    return 1 / res
print(power(float(input()), int(input())))
```

```python
def capitalize(word):
  first_letter_small = word[0]
  first_letter_big = chr(ord(first_letter_small) - ord('a') + ord('A'))
  return first_letter_big + word[1:]
source = input().split()
res = []
for word in source:
  res.append(capitalize(word))
print(' '.join(res))
```

```python
def power(a, n):
  if n == 0:
    return 1
  else:
    return a * power(a, n - 1)
print(power(float(input()), int(input())))
```

```python
def reverse():
  x = int(input())
  if x != 0:
    reverse()
  print(x)
reverse()
```

```python
def fib(n):
  if n == 1 or n == 2:
    return 1
  else:
    return fib(n - 1) + fib(n - 2)
print(fib(int(input())))
```
